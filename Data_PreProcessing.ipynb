```python
import re
```

```python
file = open('model_responses.txt', 'r+')
```

```python
doc = file.readlines()
```

```python
print(doc[:500])
```

```python
user_prompt = """Given the defects summary related to Automotive industry take out the entities as "Entity" with which type of entity it is as "Label" from data and strictly return result in json format containing ["Entity", "Label"].
"""
```

```python
sentences = []
for line in doc:
    if re.match('\s*Sentence:\s*', line):
        sentences.append(line)
print(sentences[:500])
```

```python
prompt = []
length = len('    Sentence: ')
for sentence in sentences:
    prompt.append(sentence[length:])
```

```python
prompt_new = []
for i in range(0,len(prompt),2):
    prompt_new.append(prompt[i])
```

```python
print(prompt_new[:500])
```

```python
answers = []
for line in doc:
    if line[0] == '[':
        answers.append(line)
print(answers[:500])
```
